```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * np.sin(x) - np.cos(2 * x)

def f_der(x):
    return 2 * np.cos(x) + 2 * np.sin(2 * x)
def tangent_line(z, x):
    slope = f_der(x)
    intercept = f(x) - slope * x
    return slope * z + intercept

def beauty_derivative_plot(x):
  y = 2 * np.sin(x) - np.cos(2*x)
  plt.annotate(f"f'({x:.2f}) = {f_der(x):.2f}", xy = (x, y), xytext = (x - 1.5, y - 2), color = "black", arrowprops = dict(arrowstyle = "-|>", connectionstyle = "arc3, rad = 0.5"))
  z = np.linspace(-2 * np.pi, 2 * np.pi, 500)
  plt.scatter(x, f(x), color = 'black', s = 25, linewidths = 2)
  line_1 = plt.plot(z, f(z), "b-", label = "f(x) = 2 * sin(x) - cos(2x)")
  line_2 = plt.plot(z, np.gradient(f(z), z), "r--", label = "f'(x) = 2 * cos(x) + 2 * sin(2x)")
  tangent_z = np.linspace(x - 1, x + 1, 100)
  line_3 = plt.plot(tangent_z, tangent_line(tangent_z, x), "g-", label = f"Tangent at x = {x}")
  plt.ylim(-5, 5)
  plt.legend(fontsize = 'small')
  plt.axhline(0, color = 'black', linewidth = 0.5)
  plt.axvline(0, color = 'black', linewidth = 0.5)

  plt.plot([x, x, 0], [0, y, y], 'm--')

  plt.title("Function, Derivative and Tangent Line")

  plt.xlabel('x')
  plt.ylabel('y')

  plt.grid(True)
  plt.show()
beauty_derivative_plot(-3)
beauty_derivative_plot(-2)
beauty_derivative_plot(-1.71)
```

**Find which year each species had the largest population. Mark these years in the plot of populations.**

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt("/content/drive/MyDrive/populations.txt")
years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]
max_hares_year = years[np.argmax(hares)]
max_lynxes_year = years[np.argmax(lynxes)]
max_carrots_year = years[np.argmax(carrots)]

plt.figure(figsize = (15, 10))
plt.plot(years, hares, label = 'Hares', color = 'blue')
plt.plot(years, lynxes, label = 'Lynxes', color = 'orange')
plt.plot(years, carrots, label = 'Carrots', color = 'green')

plt.scatter(max_hares_year, hares[np.argmax(hares)], color = 'blue', s = 100, edgecolor = 'black')
plt.scatter(max_lynxes_year, lynxes[np.argmax(lynxes)], color = 'orange', s = 100, edgecolor = 'black')
plt.scatter(max_carrots_year, carrots[np.argmax(carrots)], color = 'green', s = 100, edgecolor = 'black')
max_population = max(hares[np.argmax(hares)], lynxes[np.argmax(lynxes)], carrots[np.argmax(carrots)])

plt.title('The largest population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xlim(left = np.min(years), right = np.max(years))
plt.ylim(bottom = 0, top = max_population + 5000)
plt.xticks(ticks = np.arange(np.min(years) - 1, np.max(years) + 1 , 1))
plt.legend()
plt.grid()

plt.show()
```

**Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.**

```python
years, hares, lynxes, carrots = data.T

plt.bar(years, hares / (hares + lynxes), label = "Hares")
plt.bar(years, lynxes / (hares + lynxes), bottom = hares / (hares + lynxes), label = "Lynxes")
plt.bar(years, carrots / (hares + lynxes), bottom = (hares / (hares + lynxes)) + (lynxes / (hares + lynxes)), label = "Carrots")

plt.xlabel("Year")
plt.ylabel("Species")
plt.legend()

plt.show()
```

**Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.**

```python
hares_mean = np.mean(hares)
hares_std = np.std(hares)
lynxes_mean = np.mean(lynxes)
lynxes_std = np.std(lynxes)
carrots_mean = np.mean(carrots)
carrots_std = np.std(carrots)

plt.figure(figsize = (15, 10))

plt.subplot(1, 3, 1)
plt.hist(hares, bins = 10, color = 'blue', edgecolor = 'black')
plt.axvline(hares_mean, color = 'red', linestyle = '--', linewidth = 1)
plt.title('Hares Population Histogram')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend(['Mean'])

plt.subplot(1, 3, 2)
plt.hist(lynxes, bins = 10, color = 'green', edgecolor = 'black')
plt.axvline(lynxes_mean, color = 'red', linestyle = '--', linewidth = 1)
plt.title('Lynxes Population Histogram')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend(['Mean'])

plt.subplot(1, 3, 3)
plt.hist(carrots, bins = 10, color = 'orange', edgecolor = 'black')
plt.axvline(carrots_mean, color = 'red', linestyle = '--', linewidth = 1)
plt.title('Carrots Population Histogram')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.legend(['Mean'])

plt.tight_layout()
plt.show()
```

**Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value with the true value and print a ratio of the error to the true value.**

```python
var = np.polyfit(years[:-1], hares[:-1],1)
regression = np.poly1d(var)
predict = regression(1920)
truev = hares[-1]

print(f'Estimated number of hares in 1920: {predict:.2f}')
print(f'The true number of hares in 1920: {truev:.2f}')
print(f'Error rate: {abs((predict - truev) / truev):.2%}')
```
