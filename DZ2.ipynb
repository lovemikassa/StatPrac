**Задание 1.**

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * np.sin(x) - np.cos(2 * x)

def f_prime(x):
    return 2 * np.cos(x) + 2 * np.sin(2 * x)

def beauty_derivative_plot(x):
    x_values = np.linspace(-2 * np.pi, 2 * np.pi, 400)
    y_values = f(x_values)

    y_prime_values = f_prime(x_values)

    # Уравнение касательной в х
    slope = f_prime(x)

    tangent_length = 2
    x_start = x - tangent_length / 2
    x_end = x + tangent_length / 2

    y_tangent_start = slope * (x_start - x) + f(x)
    y_tangent_end = slope * (x_end - x) + f(x)

    plt.figure(1, figsize = (15, 10))

    # График f(x)
    plt.plot(x_values, y_values, label = 'f(x) = 2sin(x) - cos(2x)', color = 'blue')

    # График f'(x)
    plt.plot(x_values, y_prime_values, label = "f'(x) = 2cos(x) + 2sin(2x)", color='red', linestyle = '--')

    # Строим касательную линию в точке x
    plt.plot([x_start, x_end], [y_tangent_start, y_tangent_end], label=f'Tangent line at x = {x}', linestyle = '-', color = 'green')

    # Отмечаем точку касания
    plt.scatter([x], [f(x)], color = 'black')

    # Добавляем стрелку, указывающую на точку касания
    plt.annotate(f"f'({x:.2f}) = {slope:.2f}", xy = (x, f(x)), xytext = (x - 1.5, f(x) - 1.5), arrowprops = dict(facecolor = 'black', arrowstyle = '->'), fontsize = 10, color = 'black')

    plt.plot([x, x], [0, f(x)], color = 'purple', linestyle = '--', linewidth = 1)
    plt.plot([0, x], [f(x), f(x)], color = 'purple', linestyle = '--', linewidth = 1)

    # Добавляем подписи и легенду
    plt.title('Function, Derivative and Tangent Line')
    plt.xlabel('x')
    plt.ylabel('y')

    # Устанавливаем фиксированные пределы для осей
    plt.xlim(-6.2, 6.2)
    plt.ylim(-4, 4)

    # Настраиваем разметку
    plt.xticks(np.arange(-6, 7, 2))
    plt.yticks(np.arange(-4, 5, 1))

    # Добавляем линии осей
    plt.axhline(0, color = 'black', linewidth = 0.5)
    plt.axvline(0, color = 'black', linewidth = 0.5)

    # Настройка сетки и легенды
    plt.grid(True)
    plt.legend()

    # Показываем график
    plt.show()

# Пример использования функции
beauty_derivative_plot(-3)
beauty_derivative_plot(-2)
beauty_derivative_plot(-1.71)
```

**Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value with the true value and print a ratio of the error to the true value.**

```python
import numpy as np
import matplotlib.pyplot as plt
from google. colab import files
uploaded = files.upload()

data = np.loadtxt('populations.txt', delimiter = '	', skiprows = 1)
years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]

def linear_regression(x, y):
    n = len(x)
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    numerator = np.sum((x - x_mean) * (y - y_mean))
    denominator = np.sum((x - x_mean) ** 2)
    b = numerator / denominator
    a = y_mean - b * x_mean

    return a, b

a, b = linear_regression(years, hares)

year_to_estimate = 1920
estimated_hare_1920 = a + b * year_to_estimate

true_hare_1920 = hares[years == 1920][0]

error_ratio = (estimated_hare_1920 - true_hare_1920) / true_hare_1920

print(f'Estimated number of hares in 1920: {estimated_hare_1920:.2f}')
print(f'The true number of hares in 1920: {true_hare_1920:.2f}')
print(f'Error rate: {error_ratio:.2%}')

```
